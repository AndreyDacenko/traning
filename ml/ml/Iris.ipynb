```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
iris_path = Path.cwd() / 'data' / 'iris.data'
iris_df = pd.read_csv(iris_path, header=None)
```

```python
y = iris_df.iloc[:, 4].values ; y
```

```python
X = iris_df.iloc[:, [0,2]].values ; X
```

```python
idx0 = y == 'Iris-setosa'
idx1 = y == 'Iris-versicolor'
idx2 = y == 'Iris-virginica'
plt.scatter(X[idx0,0], X[idx0,1], color='red', marker='o', label='Setosa')
plt.scatter(X[idx1,0], X[idx1,1], color='blue', marker='x', label='Versicolor')
plt.scatter(X[idx2,0], X[idx2,1], color='green', marker='s', label='Virginica')
plt.xlabel('длина чашелистника (см)')
plt.ylabel('длина лепестка (см)')
plt.legend(loc='upper left')
```
