```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1, _ = make_blobs(n_samples=200, n_features=2, centers=4,
                   cluster_std=1.0, shuffle=True, random_state=3)
plt.scatter(X1[:,0], X1[:,1], color='red', marker='o')
```

```python
X2, _ = make_blobs(n_samples=200, n_features=2, centers=4,
                   cluster_std=1.0, shuffle=True, random_state=1)
plt.scatter(X2[:,0], X2[:,1], color='blue', marker='o')
```

# Метод k средних (k-means)

## Диаграмма Вороного

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
points = np.random.rand(5,2)
vor = Voronoi(points)
voronoi_plot_2d(vor)
```

```python
idx = np.random.randint(0,200,size=5)
idx
```

```python
X1_s = X1[idx,:]
```

```python
X1_s
```

```python
vor1 = Voronoi(X1_s)
fig = voronoi_plot_2d(vor)
ax = fig.subplots()
ax.scatter(X1[:,0], X1[:,1], color='red', marker='o')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans1 = KMeans(n_clusters=4).fit(X1)
```

```python
kmeans1.cluster_centers_
```

```python
plt.scatter(X1[:,0], X1[:,1], color='red', marker='o')
C1 = kmeans1.cluster_centers_
plt.scatter(C1[:,0],C1[:,1], color='black', marker='s')
```

```python
L1 = kmeans1.labels_
k0 = L1 == 0
k1 = L1 == 1
k2 = L1 == 2
k3 = L1 == 3
plt.scatter(X1[k0,0], X1[k0,1], color='red', marker='o')
plt.scatter(X1[k1,0], X1[k1,1], color='green', marker='o')
plt.scatter(X1[k2,0], X1[k2,1], color='blue', marker='o')
plt.scatter(X1[k3,0], X1[k3,1], color='magenta', marker='o')
```

```python
kmeans1a = KMeans(n_clusters=5).fit(X1)
L1 = kmeans1a.labels_
k0 = L1 == 0
k1 = L1 == 1
k2 = L1 == 2
k3 = L1 == 3
k4 = L1 == 4
plt.scatter(X1[k0,0], X1[k0,1], color='red', marker='o')
plt.scatter(X1[k1,0], X1[k1,1], color='green', marker='x')
plt.scatter(X1[k2,0], X1[k2,1], color='blue', marker='s')
plt.scatter(X1[k3,0], X1[k3,1], color='magenta', marker='^')
plt.scatter(X1[k4,0], X1[k4,1], color='cyan', marker='+')
```

```python
kmeans1b = KMeans(n_clusters=3).fit(X1)
L1 = kmeans1b.labels_
k0 = L1 == 0
k1 = L1 == 1
k2 = L1 == 2
plt.scatter(X1[k0,0], X1[k0,1], color='red', marker='o')
plt.scatter(X1[k1,0], X1[k1,1], color='green', marker='x')
plt.scatter(X1[k2,0], X1[k2,1], color='blue', marker='s')
```

```python
kmeans2 = KMeans(n_clusters=4).fit(X2)
L1 = kmeans2.labels_
k0 = L1 == 0
k1 = L1 == 1
k2 = L1 == 2
k3 = L1 == 3
plt.scatter(X2[k0,0], X2[k0,1], color='red', marker='o')
plt.scatter(X2[k1,0], X2[k1,1], color='green', marker='x')
plt.scatter(X2[k2,0], X2[k2,1], color='blue', marker='s')
plt.scatter(X2[k3,0], X2[k3,1], color='magenta', marker='^')
```
