```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,1000)
y = np.linspace(-10,10,1000)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
X
```

```python
Y
```

```python
Z = X + Y
```

```python
plt.contour(X,Y,Z, levels=25)
```

```python
Z = ( X**2 + Y**2 ) / 25
plt.contourf(X,Y,Z,levels=25, alpha=0.3)
plt.contour(X,Y,Z,levels=25)
```

```python
Z = np.sin((X**2+Y**2 + Y)/20)
plt.contour(X,Y,Z,levels=25)
```
